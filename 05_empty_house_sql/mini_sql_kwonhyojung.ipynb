```python
# 라이브러리 불러오기

import pymysql
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import koreanize_matplotlib
```

### df1에는 1인당 소득,경제성장률,자산,물가지수를 꺼내고
### df2에는 소득만족도와 소비 만족도도 꺼낸다.
소득만족도와 소비 만족도는 2년 단위 데이터라서 다르게 한다.

```python
# 데이터베이스에서 데이터 불러오기
conn = pymysql.connect(host='172.30.1.4',user='kwon',
                       password=YOUR_PASSWORD,database='mini',charset='utf8')
cur = conn.cursor(pymysql.cursors.DictCursor)

query="""
SELECT e.year,e.region,e.empty_rate,e.empty_number,e.total_house,pi.personal_income,c.credit,eg.economy_growth,pr.price_index
FROM empty2 as e
    INNER JOIN personal_income2 as pi
    ON e.region_id = pi.region_id
    INNER JOIN credit2 as c
    ON pi.region_id = c.region_id
    INNER JOIN economy_growth as eg
    ON pi.region_id = eg.region_id
    INNER JOIN price_index2 as pr
    ON pi.region_id = pr.region_id
    
"""
# 큰 따옴표 사용 지향(내부에 문자열('')입력 용이)
# 마지막에 세미콜론 없음
 
cur.execute(query)

rows = cur.fetchall() # 모든 데이터를 가져옴
df1 = pd.DataFrame(rows) # DataFrame 형태로 변환

```

```python
# 데이터베이스에서 데이터 불러오기
query2="""
SELECT e.year,e.region,e.empty_rate,pi.personal_income,cs.consumer5,cs.consumer4,c.credit,eg.economy_growth,ins.income5,ins.income4,pr.price_index
FROM empty2 as e
    INNER JOIN personal_income2 as pi
    ON e.region_id = pi.region_id
    INNER JOIN consumer_satisfaction as cs
    ON pi.region_id = cs.region_id
    INNER JOIN credit2 as c
    ON pi.region_id = c.region_id
    INNER JOIN economy_growth as eg
    ON pi.region_id = eg.region_id
    INNER JOIN income_satisfaction as ins
    ON pi.region_id = ins.region_id
    INNER JOIN price_index2 as pr
    ON pi.region_id = pr.region_id
    
"""
# 큰 따옴표 사용 지향(내부에 문자열('')입력 용이)
# 마지막에 세미콜론 없음
 
cur.execute(query2)

rows2 = cur.fetchall() # 모든 데이터를 가져옴
df2 = pd.DataFrame(rows2) # DataFrame 형태로 변환
```

```python
# df1과 df2 데이터 확인
display(df1)
display(df2)
```

```python
# df2의 소비만족도,소득만족도를 float 데이터로 변환
df2 = df2.astype({'consumer5':'float','consumer4':'float','income5':'float','income4':'float'})
df2['income_good'] = df2['income5'] + df2['income4']
df2['consumer_good'] = df2['consumer5'] + df2['consumer4']
print(df1.info())
print(df2.info())
```

### 2015~2024년까지 region에 대해 평균을 내서 데이터를 시각화

```python
# 2015~2024년까지 region에 대해 평균을 내서 데이터를 시각화

df3 = df1.groupby('region')[['empty_rate','personal_income','credit','economy_growth','price_index']].mean().sort_values(by='empty_rate')
plt.figure(figsize=(12,8))
bars = plt.bar(df3.index,df3['empty_rate'])
# 2. 숫자 넣기 (반복문 사용)
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height,
             f'{height:.1f}',
             ha='center', va='bottom') # ha: 가로 정렬, va: 세로 정렬
plt.title('2015~2024 전국 빈집 비율 차트',size=20)
plt.xticks(rotation=90)
plt.show()
```

```python
# 1인당 소득과 전국 빈집 비율 차트
fig,ax1=plt.subplots()
df3.plot(kind='bar',y='empty_rate',ax=ax1)
ax2=ax1.twinx()
df3.plot(y='personal_income',color='red',marker='o',ax=ax2)
plt.title('1인당 소득과 전국 빈집 비율 차트',size=20)
plt.xticks(rotation=90)
plt.show()
```

```python
# 소비자 물가지수와 전국 빈집
fig,ax1=plt.subplots()
df3.plot(kind='bar',y='empty_rate',ax=ax1)
ax2=ax1.twinx()
df3.plot(y='price_index',color='red',marker='o',ax=ax2)
plt.title('소비자 물가지수와 전국 빈집 비율 차트',size=20)
plt.xticks(rotation=90)
plt.show()
```

```python
fig,ax1=plt.subplots()
df3.plot(kind='bar',y='empty_rate',ax=ax1)
ax2=ax1.twinx()
df3.plot(y='economy_growth',color='red',marker='o',ax=ax2)
plt.title('경제성장률과 전국 빈집 비율 차트',size=20)
plt.xticks(rotation=90)
plt.show()
```

```python
# 예금액 전국은 모든 것을 더한 값이기 때문에 여기서는 뺀다. 
fig,ax1=plt.subplots()
df3.drop(index='전국').plot(kind='bar',y='empty_rate',ax=ax1)
ax2=ax1.twinx()
df3.drop(index='전국').plot(y='credit',color='red',marker='o',ax=ax2)
plt.title('예금액과 전국 빈집 비율 차트',size=20)
plt.xticks(rotation=90)
plt.show()
```
